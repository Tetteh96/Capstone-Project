```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
sales_data = pd.read_csv('sales_2019.csv')
sales_data = sales_data.rename(columns={'Order ID' : 'Order_ID', 'Price Each' : 'Price', 'Order Date' : 'Date',
                                'Quantity Ordered' : 'Quantity', 'Purchase Address' : 'Purchase_Address'})

sales_data['Date'] = pd.to_datetime(sales_data['Date']).dt.date
sales_data['City'] = sales_data['Purchase_Address'].str.split(',').str[1].str.strip()
sales_data = sales_data.drop(columns='Purchase_Address')
sales_data
```

```python
sales_data.to_csv('cleaned_data.csv', index=False)
```
